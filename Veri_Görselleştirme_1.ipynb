```python
import matplotlib.pyplot as plt
```

## Veri seti tanımlama 


```python
x = [1, 2, 3, 4, 5, 6, 7, 8]
y = [1, 4, 9, 16, 25, 36, 49, 64]
```

# Çizgi grafiği için gereken kod


```python
plt.plot(x,y)
plt.xlabel("x ekseni")
plt.ylabel("y ekseni")
plt.show()
```

## nokta grafiği için yapılması gereken değişiklik "."*


```python
plt.plot(x,y, ".")
plt.xlabel("x ekseni")
plt.ylabel("y ekseni")
plt.show()
```

## Scatter (daha fazla veri gerekir dağılım grafiği)

```python
plt.scatter(x,y)
plt.show()
```

# Bar Grafiği

```python
a = ["Hasta", "saglıklı", "deli"]
b = [300, 800, 500]
plt.bar(a, b)
plt.plot(a, b, color="gray")
plt.show()
```

#WorkShop


```python
x = [1, 2, 3, 4, 5, 6, 7, 8]
y = [1, 4, 9, 16, 25, 36, 49, 64]

x2 = [1, 2, 3, 4, 5, 6, 7, 8]
y2 = [1, 8, 27, 64, 125, 216, 343, 512]
```

```python
plt.plot(x, y, label="X Kare", color="red")
plt.plot(x2, y2, label="X Küp", color="purple")


plt.title("Line Plot")
plt.xlabel("X Ekseni")
plt.ylabel("Y Ekseni")


plt.legend()
plt.grid()
plt.show()
```

#SubPlots


```python
fig = plt.figure(figsize=(18,5))

first_plot = fig.add_subplot(1,3,1)
first_plot.plot(x,y, color="red")
first_plot.set_title("one")

second_plot = fig.add_subplot(1,3,2)
second_plot.scatter(x,y, color="green")
second_plot.set_title("two")

third_plot = fig.add_subplot(1,3,3)
third_plot.bar(x,y, color="darkblue")
third_plot.set_title("three")

plt.show()
```
